```python
from sklearn.model_selection import GridSearchCV

params = {
    'criterion':  ['gini', 'entropy'],
    'max_depth':  [None, 2, 4, 6, 8, 10],
    'max_features': [None, 'sqrt', 'log2', 0.2, 0.4, 0.6, 0.8],
    'splitter': ['best', 'random']
}

clf = GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid=params,
    cv=5,
    n_jobs=5,
    verbose=1,
)
clf.fit(X_train, y_train)
print(clf.best_params_)

```

part 2

```python
apears = pd.read_csv('/kaggle/input/sajazhr/Hotel Reservations.csv',index_col="Booking_ID")
y=apears.pop("booking_status").values
X=apears.values
print(X)
print(y)
apears.head()
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer


# use make_column_transformer and OneHotEncoder to convert categorical features to numbers
column_transformer = make_column_transformer((OneHotEncoder(), ['type_of_meal_plan', 'room_type_reserved', 'market_segment_type']), remainder='passthrough')
X_transformed = column_transformer.fit_transform(apears.iloc[:, :-1])
X_transformed = pd.DataFrame(data=X_transformed)


# split dataset to train/test split
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.2, random_state=100)

```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

predictions = clf.predict(X_test)
print(accuracy_score(y_test, predictions))
```

How accurate is the model? good
Does the model improve by using categorical features? yes
